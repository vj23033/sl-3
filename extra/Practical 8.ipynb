```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/titanic_data.csv')
```

```python
data
```

```python
data.describe()
```

```python
data.shape
```

```python
data.describe(include = "object")
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(np.mean(data['Age']))
```

```python
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data
```

```python
sns.countplot(data['Pclass'])
```

```python
sns.countplot(data['Survived'])
```

```python
sns.histplot(data['Sex'])
```

```python
sns.histplot(data['Embarked'])
```

```python
sns.catplot(x='Pclass',y='Age',data=data,kind='box')
```

```python
sns.catplot(x=data['Pclass'],y=data['Fare'],data=data,kind='strip')
```

```python
sns.catplot(x=data['Sex'],y=data['Fare'],data=data,kind='strip')
```

```python
sns.catplot(x=data['Sex'],y=data['Age'],data=data,kind='strip')
```

```python
sns.pairplot(data)
```

```python
sns.scatterplot(x=data['Fare'],y=data['Pclass'],hue='Survived',data=data)
```

```python
sns.scatterplot(x=data['Survived'],y=data['Fare'],data=data)
```

```python
sns.distplot(data['Age'])
```

```python
sns.distplot(data['Fare'])
```

```python
sns.jointplot(x=data['Survived'],y=data['Fare'],kind='scatter',data=data)
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])

tc = numeric_data.corr()

sns.heatmap(tc, cmap="YlGnBu")
plt.title('Correlation')
plt.show()
```

```python
sns.catplot(x='Pclass', y='Fare', data=data, kind='bar')
```
